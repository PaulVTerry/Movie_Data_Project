# Movie Data Graphs

# Step 1: Importing Data.

```python
import matplotlib.pyplot as plt
import pandas as pd
import csv

df = pd.read_csv("tmdb-movies.csv")
```

# Step 2: Data Cleaning

In the code below I wanted to return 'revenue' and 'budget' data with values greater than zero. 

```python
print(len(df))  #returns number of elements in uncleaned/raw data frame

df1 = df[df["revenue"] > 0]
df2 = df1[df1["budget"] > 0]

print(len(df2)) #returns number of elements in data frame with values over zero in 'budget' & 'revenue' columns
```

In the code's output above, we can see the attempt at data cleaning was effective because the new number (3855) of elements returned is less than the original number (10866) of elements returned, implying our code was successful at excluding values less than or equal to zero.

## Function 1

Sorts data by 'popularity' with respect to 'original_title'

With this function we are able to sort data according to popularity.

```python
def sort_by_gross_profit(df2):
    return df2.sort_values(by = 'popularity',ascending = False)['original_title']
sort_by_gross_profit(df2)
```

# Step 3: Asking Questions.

## Question 1: Which movie is the most popular?

## Bar Graph
### Graph 1.1

```python
df3 = df2.head(20) #stores the first ten values of df2 for easy graphing.
```

The bar graph below represents 'popularity' as a function of 'original_title' and shows us which movie is the most popular.

```python
plt.bar(df3["original_title"], df3["popularity"], color="blue", label="Bar graph")
plt.scatter(df3["original_title"], df3["popularity"], color="orange", label="Scatter")
plt.legend()
plt.xlabel("Movie Title", color="blue")
plt.xticks(rotation=90)
plt.ylabel("Popularity", color="blue")
plt.title("Movie Popularity", color="blue")
plt.xlim(-1,len(df3)) #x plane adjusts to the length of the list
plt.ylim(0,35)
plt.show()
```

## Scatter Plot
### Graph 1.2

The scatter plot below is another way to graphically represent popularity as a function of the movie title. This graph is intended to demonstrate matplotlibs scatterplot functionality and my understanding of it.

```python
plt.plot(df3["original_title"], df3["popularity"], color="orange", label="Line Graph")
plt.scatter(df3["original_title"], df3["popularity"], color="red", label = "Scatter Plot")
plt.legend()
plt.xlabel("Movie Title", color="red")
plt.xticks(rotation=90)
plt.ylabel("Popularity", color="red")
plt.title("Movie Popularity", color="red")
plt.xlim(-1,len(df3))
plt.ylim(0,35)
plt.show()
```

### Graph 1.3

An alternate representation of 'popularity' as a function of 'original_title'

```python
plt.barh(df3["original_title"], df3["popularity"], color=["red", "green", "blue"], alpha= .5)
plt.ylabel("Movie Title", color="blue")
plt.xlabel("Popularity", color="blue")
plt.title("Movie Popularity", color="blue")
plt.show()
```

# Pie Chart

The pie chart below shows how many top 20 movies have popularity ratings; over or equal to 25, less than 25 and greater or equal to 10, and less than 10. The first two figures are the the same as the figures above. I've included a pie chart to demonstrate matplotlibs functionality and my understanding of it.

```python
fig=plt.figure()
fig.patch.set_facecolor("lightgrey")
graph1 = fig.add_subplot(2,2,1, facecolor='lightyellow')
graph1.plot(df2["popularity"], df2["budget"])
graph1.set_title("Budget vs Popularity")

graph2 = fig.add_subplot(2,2,2, facecolor='lightblue')
graph2.plot(df2["popularity"], df2["revenue"])
graph2.set_title("Revenue vs Popularity")

graph3 = fig.add_subplot(2,1,2, facecolor='lightgreen')

x = len(df3[df3.popularity>=25])
x1 = len( df3[ (df3.popularity>=10) & (df3.popularity<25)])
x3 = len(df3[df3.popularity<10])
graph3.axis('equal')
graph3.pie([x,x1,x3],colors=['yellow','blue','green'], labels=['>=25','<25 & >=10','<10'])
graph3.legend(title="Popularity")
plt.show()
```
